```python
folium_colors = ['red', 'green', 'purple', 'orange', 'darkred', 'blue', 'black',
             'cadetblue', 'darkblue', 'lightblue', 'black', 'darkgreen', 
             'gray', 'lightgreen',
             'lightgray', 'beige', 'white'] # darkpurple, lightred

distritos['Departamento'] = pd.Categorical(distritos['Departamento'])
almacenes['Departamento'] = pd.Categorical(almacenes['Departamento'])
```

```python
def draw_map(df, popup_col, colors):
    m = folium.Map(
        location=[-13.584810,-74.524353],
        zoom_start=6,
        tiles='openstreetmap') # openstreetmap stamenterrain

    m.choropleth(
        geo_data='data/peru_departamental_simple.geojson',
        name='choropleth',
        fill_color='lightblue',
        fill_opacity=0.3,
        line_opacity=1,
    )

    folium.LayerControl().add_to(m)

    for idx, row in df.iterrows():
#         print(df.Departamento.cat.codes[idx])
        folium.CircleMarker([row.Latitud, row.Longitud], radius=3,
                popup=f'<b>{row[popup_col]}</b>', 
                color=folium_colors[df.Departamento.cat.codes[idx]]).add_to(m)
    return m
```

```python
draw_map(distritos, 'Distrito', folium_colors)
```

```python
draw_map(almacenes, 'DESCRIPCION', folium_colors)
```
