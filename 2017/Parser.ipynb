```python
def preprocess(df):
    header  = np.array(df.columns)
    header0 = np.array(df.iloc[0])
    header1 = np.array(df.iloc[1])
    header2 = np.array(df.iloc[2])
    columns = header2
    columns = [col1 if col1 == col1 else col2 for col1, col2 in zip(columns, header1)]
    columns = [col1 if col1 == col1 else col2 for col1, col2 in zip(columns, header0)]
    columns = [col1 if col1 == col1 else col2 for col1, col2 in zip(columns, header)]
    df = df.drop(df.index[0])
    df = df.drop(df.index[0])
    df = df.drop(df.index[0])
    df.columns = columns
    return df    
```

```python
df_heladas = preprocess(df_heladas)
df_friaje = preprocess(df_friaje)
```

### Latitude and longitude Checkpoint

```python
df = pd.read_csv('latitud+longitud+completo.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df
```

```python
df['DISTRITO'] = df['DISTRITO'].apply(lambda x: x.replace('\r\n', ' '))
```

```python
locations = [
    {
        'tambo': {
            'nombre': 'Tambo Aymaña',
            'distrito': 'Aymaña',
            'provincia': 'Carabaya',
            'departamento': 'Puno',
            'latitud': -15,
            'longitud': 70,
            'poblacion': 1000
        },
        'distritos': {
            'distrito': 'Aymaña',
            'provincia': 'Carabaya',
            'departamento': 'Puno',
            'latitud': -15,
            'longitud': 70,
            'poblacion': 100
        }
    }
]
```

```python
def make_dict(row):
    return {
        "distrito": row['Distrito'],
        "provincia": row['Provincia'],
        "departamento": row['Departamento'],
        "latitud": row['Latitud'],
        "longitud": row['Longitud'],
        "poblacion": row['Total del distrito'],
    }
```

```python
locations = []
for tambo in tambos:
    distritos = df[df['LOCAL DE DISTRIBUCIÓN DE BIENES'] == tambo]
    distrito_tambo = distritos[distritos['Distrito'] == distritos['DISTRITO']]['Distrito']
    distrito_tambo = distrito_tambo.iloc[0] if len(distrito_tambo) > 0 else distritos['Distrito'].iloc[0]
    
    location = {
        "tambo": {
            "nombre": tambo,
            "distrito": distritos['DISTRITO'].unique()[0],
            "provincia": distritos['Provincia'].unique()[0],
            "departamento": distritos['Departamento'].unique()[0],
            "latitud": distritos[distritos['Distrito'] == distrito_tambo]['Latitud'].iloc[0],
            "longitud": distritos[distritos['Distrito'] == distrito_tambo]['Longitud'].iloc[0],
            "poblacion": int(distritos['Total del distrito'].sum())
        },
        "distritos": list(map(lambda x: make_dict(x), distritos.to_dict('records')))
    }
    
    locations += [location]
```

```python
locations
```

```python
import json
with open('locations.json', 'w', encoding="latin-1") as outfile:
    json.dump(locations, outfile)
```
