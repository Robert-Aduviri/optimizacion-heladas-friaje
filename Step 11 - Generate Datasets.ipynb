## Run models

```python
from utils import load_file
E, D, W, d, g, v, t, a, c = load_file('data/test1.txt')
```

```python
from utils import chromosome_fitness
from utils import init_genetic_algorithm
from deap import tools, algorithms

toolbox, stats = init_genetic_algorithm(chromosome_fitness,
                            E, D, W, d, g, v, t, a, c)
pop = toolbox.population(n=100)
hof = tools.HallOfFame(5)
```

```python
pop, log = algorithms.eaSimple(pop, toolbox, cxpb=0.5, mutpb=0.2,
                              ngen=30, stats=stats, halloffame=hof,
                              verbose=True)
```

```python
from utils import plot_log
plot_log(log)
```

## Plot times

```python
ga_max_val = [12130, 18400, 22520, 33020, 21790, 41390]
ga_min_val = [ 8000, 10880, 13150, 20470, 19480, 28080]
ga_time    = [270, 328, 419, 660, 927, 1180]
glpk_val   = [ 8000, 1.08e4, 1.304e4, 1.942e4, 1.893e4, 2.669e4]
glpk_time  = [113, 113, 133, 594, 1*1e3+740, 3*60*1e3+46*1e3]
size       = [10, 15, 20, 30, 40, 50]
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(size[:], ga_time[:], label='Genetic Algorithm')
ax.plot(size[:-1], glpk_time[:-1], label='GLPK solver')
plt.legend()
plt.title('Execution time vs input size')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(size, ga_max_val, label='Genetic Algorithm (max value)')
ax.plot(size, ga_min_val, label='Genetic Algorithm (min value)')
ax.plot(size, glpk_val, label='GLPK solver (optimal value)')
plt.legend()
plt.title('Objective value vs input size')
```
