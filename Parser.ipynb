```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_heladas = pd.read_excel('DISTRITOS_PRIORIZADOS-POR-HELADAS.xlsx')
df_friaje = pd.read_excel('DISTRITOS_PRIORIZADOS-POR-FRIAJE.xlsx')
```

```python
def preprocess(df):
    header  = np.array(df.columns)
    header0 = np.array(df.iloc[0])
    header1 = np.array(df.iloc[1])
    header2 = np.array(df.iloc[2])
    columns = header2
    columns = [col1 if col1 == col1 else col2 for col1, col2 in zip(columns, header1)]
    columns = [col1 if col1 == col1 else col2 for col1, col2 in zip(columns, header0)]
    columns = [col1 if col1 == col1 else col2 for col1, col2 in zip(columns, header)]
    df = df.drop(df.index[0])
    df = df.drop(df.index[0])
    df = df.drop(df.index[0])
    df.columns = columns
    return df    
```

```python
df_heladas = preprocess(df_heladas)
df_friaje = preprocess(df_friaje)
```

### Begin merge

**Replace line break symbols:**

```python
df_tambos['Departamento'] = df_tambos['Departamento'].apply(lambda x: x.replace('\n', ' '))
df_tambos['Provincia'] = df_tambos['Provincia'].apply(lambda x: x.replace('\n', ' '))
df_tambos['Distrito'] = df_tambos['Distrito'].apply(lambda x: x.replace('\n', ' '))
```

```python
df_heladas['Departamento'] = df_heladas['Departamento'].apply(lambda x: x.replace('\n', ' '))
df_heladas['Provincia'] = df_heladas['Provincia'].apply(lambda x: x.replace('\n', ' '))
df_heladas['Distrito'] = df_heladas['Distrito'].apply(lambda x: x.replace('\n', ' '))
```

```python
df_friaje['Departamento'] = df_friaje['Departamento'].apply(lambda x: x.replace('\n', ' '))
df_friaje['Provincia'] = df_friaje['Provincia'].apply(lambda x: x.replace('\n', ' '))
df_friaje['Distrito'] = df_friaje['Distrito'].apply(lambda x: x.replace('\n', ' '))
```

**Concat df_heladas and df_friaje:**

```python
set(df_heladas.columns).symmetric_difference(df_friaje.columns)
```

```python
df_heladas.shape, df_friaje.shape
```

```python
df_heladas.append(df_friaje).shape
```

```python
df = pd.merge(df_tambos, df_heladas.append(df_friaje), how='left', on=['Departamento', 'Provincia', 'Distrito'])
```

```python
df.to_csv('merge.csv', index=False)
```
