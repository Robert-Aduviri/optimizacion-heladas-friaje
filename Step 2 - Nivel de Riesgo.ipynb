```python
common_columns = ['Ubigeo distrital', 'Departamento', 'Provincia', 'Distrito',
                    'T° Mínima P10 (°C)', 'Frecuencia de heladas (días/mes)']
```

### 1. Factores más importantes para determinar el nivel de riesgo según MINSA

```python
from lightgbm import LGBMRegressor, plot_importance
```

```python
label_dict = {
    'BAJO': 1,
    'MEDIO': 2,
    'ALTO': 3,
    'MUY ALTO': 4
}
```

```python
df_salud['Nivel de Riesgo MINSA'] = df_salud['Nivel de Riesgo MINSA'].apply(label_dict.get)
```

```python
for col in df_salud.columns:
    if col not in common_columns:
        df_salud[col] = df_salud[col].astype(float)
```

```python
lgbm = LGBMRegressor()
lgbm.fit(df_salud.drop(common_columns + ['Nivel de Riesgo MINSA'], axis=1), df_salud['Nivel de Riesgo MINSA'])
```

```python
plot_importance(lgbm)
```
